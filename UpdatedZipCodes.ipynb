```python
import plotly.graph_objects as go
import numpy as np
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv('zip_data.csv')
df.head()
```

```python
def fun2020to2021(row):
    if (row['2019-2020'] == 0) and (row['2020-2021'] == 0):
        return 0
    elif (row['2019-2020'] == 0) and (row['2020-2021'] != 0):
        return 100
    elif (row['2019-2020'] != 0):
        return ((row['2020-2021'] - row['2019-2020'])/row['2019-2020'])*100
    
def fun2021to2022(row):
    if (row['2020-2021'] == 0) and (row['2021-2022'] == 0):
        return 0
    elif (row['2020-2021'] == 0) and (row['2021-2022'] != 0):
        return 100
    elif (row['2020-2021'] != 0):
        return ((row['2021-2022'] - row['2020-2021'])/row['2020-2021'])*100
    
def fun2022to2023(row):
    if (row['2021-2022'] == 0) and (row['2022-2023'] == 0):
        return 0
    elif (row['2021-2022'] == 0) and (row['2022-2023'] != 0):
        return 100
    elif (row['2021-2022'] != 0):
        return ((row['2022-2023'] - row['2021-2022'])/row['2021-2022'])*100
    
def fun2023to2024(row):
    if (row['2022-2023'] == 0) and (row['2023-2024'] == 0):
        return 0
    elif (row['2022-2023'] == 0) and (row['2023-2024'] != 0):
        return 100
    elif (row['2022-2023'] != 0):
        return ((row['2023-2024'] - row['2022-2023'])/row['2022-2023'])*100
```

```python

df['2020-2021 Change'] = df.apply(fun2020to2021, axis = 1)
df['2021-2022 Change'] = df.apply(fun2021to2022, axis = 1)
df['2022-2023 Change'] = df.apply(fun2022to2023, axis = 1)
df['2023-2024 Change'] = df.apply(fun2023to2024, axis = 1)

df['2020-2021 Change'] = df['2020-2021 Change'].round(2)
df['2021-2022 Change'] = df['2021-2022 Change'].round(2)
df['2022-2023 Change'] = df['2022-2023 Change'].round(2)
df['2023-2024 Change'] = df['2023-2024 Change'].round(2)
df.head(20)
```

```python
# Total cash disbursments 2019-2024 by zip code, unweighted

fig = px.scatter_mapbox(df,
    mapbox_style = 'open-street-map',
    lon = 'Lon',
    lat = 'Lat',
    opacity = 0.75,
    title = '<b>Total Cash Disbursments 2019-2024 by ZIP Code<b>',
)

fig.update_traces(hovertemplate="Zip Code: " + df['ZIP'].astype(str) + "<br>Total: $" + df['Total'].astype(str))
fig.show()
#fig.write_html('Total_Disbursments_2019-2024.html')
```

```python
# 2019-2020, weighted

fig = px.scatter_mapbox(df,
    mapbox_style = 'open-street-map',
    lon = 'Lon',
    lat = 'Lat',
    opacity = 0.75,
    size = df['2019-2020']/100,
    title = '<b>Total Cash Disbursments 2019-2020 by ZIP Code, Weighted<b>',
)

fig.update_traces(hovertemplate="Zip Code: " + df['ZIP'].astype(str) + "<br>Total: $" + df['2019-2020'].astype(str))
fig.show()
#fig.write_html('Total_Disbursments_2019-2020.html')
```
